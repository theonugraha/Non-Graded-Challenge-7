Carilah derivative dari sebuah fungsi:


y = x^2+2x+1


y = 4x^3 - 3x^2 + 2x - 1

```python
import sympy as sy
```

```python
x = sy.symbols('x')
```

### y = x^2+2x+1

```python
y = x**2 + 2*x + 1
y
```

```python
y.diff() # Turunan Pertama
```

```python
y.diff().diff() # Turunan Kedua
```

```python
y.diff().diff().diff() # Turunan Ketiga
```

### y = 4x^3 - 3x^2 + 2x - 1

```python
y = 4*x**3 - 3*x**2 + 2*x - 1
y
```

```python
y.diff() # Turunan Pertama
```

```python
y.diff().diff() # Turunan Kedua
```

```python
y.diff().diff().diff() # Turunan Ketiga
```

```python
y.diff().diff().diff().diff() # Turunan Keempat
```
